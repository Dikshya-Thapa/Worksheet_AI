Problem 4.1: Array Creation

```python
import numpy as np
import time

# Problem 1: Array Creation

# 1. Initialize an empty array with size 2X2.
empty_2x2 = np.empty((2, 2))
print("1) Empty 2x2 array:\n", empty_2x2, "\n")

# 2. Initialize an all one array with size 4X2.
ones_4x2 = np.ones((4, 2))
print("2) Ones 4x2 array:\n", ones_4x2, "\n")

# 3. Return a new array of given shape and type, filled with fill value.
# Example: shape (3,3), fill value = 7
full_array = np.full((3, 3), 7)
print("3) Full array 3x3 filled with 7:\n", full_array, "\n")

# 4. Return a new array of zeros with same shape and type as a given array.
given = np.array([[1, 2, 3], [4, 5, 6]])
zeros_like_given = np.zeros_like(given)
print("4) Zeros like given array:\n", zeros_like_given, "\n")

# 5. Return a new array of ones with same shape and type as a given array.
ones_like_given = np.ones_like(given)
print("5) Ones like given array:\n", ones_like_given, "\n")

# 6. Convert list new_list = [1,2,3,4] to numpy array.
new_list = [1, 2, 3, 4]
arr_from_list = np.array(new_list)
print("6) Array from list:", arr_from_list)
```

Problem 3: Array Operations

```python
# Problem 3: Array Operations

x = np.array([[1, 2],
              [3, 5]])
y = np.array([[5, 6],
              [7, 8]])

v = np.array([9, 10])
w = np.array([11, 12])

# 1. Add x and y.
add_xy = x + y
print("1) x + y:\n", add_xy, "\n")

# 2. Subtract x and y.
sub_xy = x - y
print("2) x - y:\n", sub_xy, "\n")

# 3. Multiply array with integer of choice (say 3).
mul_x = x * 3
print("3) x * 3:\n", mul_x, "\n")

# 4. Square of each element.
square_x = x ** 2
print("4) x squared:\n", square_x, "\n")

# 5. Dot products:
dot_vw = np.dot(v, w)      # scalar
dot_xv = np.dot(x, v)      # result shape (2,)
dot_xy = np.dot(x, y)      # result shape (2,2)

print("5a) v · w =", dot_vw)
print("5b) x · v =", dot_xv)
print("5c) x · y:\n", dot_xy, "\n")

# 6. Concatenate:
# x and y along rows (axis=0)
concat_xy_rows = np.concatenate((x, y), axis=0)

# v and w along columns => make them column vectors first
concat_vw_cols = np.concatenate((v.reshape(-1, 1), w.reshape(-1, 1)), axis=1)

print("6a) concat x & y along rows:\n", concat_xy_rows)
print("6b) concat v & w along columns:\n", concat_vw_cols, "\n")

# 7. Concatenate x and v and explain error.
try:
    concat_xv = np.concatenate((x, v))
    print("7) concat(x, v):\n", concat_xv)
except Exception as e:
    print("7) concat(x, v) ERROR because shapes differ.")
    print("   x shape:", x.shape, " (2D)")
    print("   v shape:", v.shape, " (1D)")
    print("   Error:", e)
```

Problem 4: Matrix Operations + Linear System

```python
# Problem 4: Matrix Operations

A = np.array([[3, 4],
              [7, 8]])
B = np.array([[5, 3],
              [2, 1]])

# 1. Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
proof_identity = A @ A_inv
print("1) A @ inv(A):\n", proof_identity, "\n")  # should be close to identity

# 2. Prove AB != BA
AB = A @ B
BA = B @ A
print("2) AB:\n", AB)
print("   BA:\n", BA)
print("   AB == BA ?", np.allclose(AB, BA), "\n")

# 3. Prove (AB)^T = B^T A^T
lhs = (AB).T
rhs = B.T @ A.T
print("3) (AB)^T:\n", lhs)
print("   B^T A^T:\n", rhs)
print("   Equal ?", np.allclose(lhs, rhs), "\n")

# ---- Solve system of equations using inverse method ----
# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9

coeff = np.array([
    [2, -3,  1],
    [1, -1,  2],
    [3,  1, -1]
])

const = np.array([-1, -3, 9])

# Solve using inverse: X = inv(A) @ B
solution_inv = np.linalg.inv(coeff) @ const
print("Solution using inverse:", solution_inv)

# Solve using np.linalg.solve (more stable)
solution_solve = np.linalg.solve(coeff, const)
print("Solution using solve:", solution_solve)
```

Problem 4.2: How Fast is NumPy?

```python
# Problem 4.2: How Fast is NumPy?

size = 1_000_000

# Python lists
L1 = list(range(size))
L2 = list(range(size))

# NumPy arrays
A1 = np.arange(size)
A2 = np.arange(size)

# 1) Element-wise Addition
start = time.time()
list_add = [L1[i] + L2[i] for i in range(size)]
list_t_add = time.time() - start

start = time.time()
numpy_add = A1 + A2
numpy_t_add = time.time() - start

print("1) Addition:")
print("Python:", list_t_add)
print("NumPy :", numpy_t_add, "\n")

# 2) Multiplication
start = time.time()
list_mul = [L1[i] * L2[i] for i in range(size)]
list_t_mul = time.time() - start

start = time.time()
numpy_mul = A1 * A2
numpy_t_mul = time.time() - start

print("2) Multiplication:")
print("Python:", list_t_mul)
print("NumPy :", numpy_t_mul, "\n")

# 3) Dot Product
start = time.time()
list_dot = sum(L1[i] * L2[i] for i in range(size))
list_t_dot = time.time() - start

start = time.time()
numpy_dot = np.dot(A1, A2)
numpy_t_dot = time.time() - start

print("3) Dot Product:")
print("Python:", list_t_dot)
print("NumPy :", numpy_t_dot, "\n")

# 4) Matrix Multiplication 1000x1000
mat_size = 1000

# Generate data
M1_list = np.random.random((mat_size, mat_size)).tolist()
M2_list = np.random.random((mat_size, mat_size)).tolist()

# Python matmul
start = time.time()
M2_T = list(zip(*M2_list))
M_result = [[sum(a*b for a,b in zip(row,col)) for col in M2_T] for row in M1_list]
list_t_mat = time.time() - start

# NumPy matmul
M1_np = np.array(M1_list)
M2_np = np.array(M2_list)

start = time.time()
M_np_res = M1_np @ M2_np
numpy_t_mat = time.time() - start

print("4) Matrix Multiplication:")
print("Python:", list_t_mat)
print("NumPy :", numpy_t_mat)
```
